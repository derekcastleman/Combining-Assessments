# SBAC and CAST Assessment Combine

The purpose of this Jupyter notebook is to combine the data from multiple SBAC and CAST assessments on the Student ID. This should allow for easier analysis of any of the data from the SBAC and CAST assessments across up to four tests. The DFS for each student will be calculated and the change across years will be calculated.

```python
import numpy as np
import pandas as pd
```
